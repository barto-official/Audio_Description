#Benchmarking different models — Checking time

This modified script removes the part of attaching srt to your video because this is independent of models and takes the most time. Thus, wihtout it we can easily compare speed based on the different model sizes (if you want the total time of script, you can just add the constant from attaching srt to mp4 because that's regardless of the model you choose).

Let's run tests for tiny, small, medium, and large models.

We use mainly two models:
1. Whisperx for transcription.
2. Pre-trained and fine-tuned Whisper for audio-captoning.

For testing we use video of 11 minutes.

**Results:**
1. Tiny: 188.44s
2. Small: 350s
3. Medium: 284s
4. Large: Too much memory overhead to handle using basic Colab plan.
